```python
import pandas as pd
import numpy as np
import itertools
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

df = pd.read_csv(r'angle_len_data.csv')
print(df)
```

```python
frames = np.zeros(len(df['n']))
frame = -1

i = 0
for n in df['n']:
    if n == 0:
        frame = frame + 1
    frames[i] = frame
    i = i + 1    

df['frame'] = frames
  
# Observe the result
print(df)
```

```python
sns.set_theme(style="whitegrid")

# my_pal = {root: "darkolivegreen" if (root == 1) else "honeydew" for 
#     root in df_1.DM}

sns.lineplot(data=df, x="frame", y='angle', hue='n')
```
