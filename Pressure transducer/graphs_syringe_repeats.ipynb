```python
import pandas as pd
import numpy as np
import itertools
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'100ml_data.csv')
print(df)
```

```python
df['pressure 1'] = df['pressure 1']*0.133333
df['pressure 2'] = df['pressure 2']*0.133333
df['pressure 3'] = df['pressure 3']*0.133333
df['pressure 4'] = df['pressure 4']*0.133333

df['pressure/vacuum 1'] = df['pressure/vacuum 1']*0.133333
df['pressure/vacuum 2'] = df['pressure/vacuum 2']*0.133333
df['pressure/vacuum 3'] = df['pressure/vacuum 3']*0.133333
df['pressure/vacuum 4'] = df['pressure/vacuum 4']*0.133333
```

```python
sns.set_theme(style="whitegrid")

ax = sns.lineplot(data=df, x="index 1", y='pressure/vacuum 1', hue='reading 1')
ax.set(xlabel='Sample Rate', ylabel='Pressure (kPa)')
plt.show()
```
