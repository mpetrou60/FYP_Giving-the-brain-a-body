```python
# Independent T-Tests
# Independent Samples T Tests Hypotheses
# If the p-value is less than your significance level (e.g., 0.05), you can reject the null hypothesis. 
# The difference between the two means is statistically significant. 
# Your sample provides strong enough evidence to conclude that the two population means are not equal.
from scipy import stats
m1m2_12_ttest = stats.ttest_ind(m2_12v, m1_12v)
m1m2_10_ttest = stats.ttest_ind(m2_10v, m1_10v)

print(m1m2_12_ttest)
print(m1m2_10_ttest)
```
